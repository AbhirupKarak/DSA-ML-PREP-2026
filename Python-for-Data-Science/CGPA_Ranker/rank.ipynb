```python
import numpy as np
import pandas as pd

#First I downloaded results pdf, then converted it to csv
```

```python
df = pd.read_csv("data_science/final_cleaned_results_sem3.csv")
```

```python
df = df[df['remarks'].isin(['PASSED','SUPP'])] #Cleaning up initial csv file
```

```python
df.to_csv('new_results.csv', index = False)
```

```python
df['rank'] = df['cgpa'].rank(method="dense", ascending = False).astype(int)
```

```python
rank_df = df[['roll_no','program','name','cgpa','rank']].sort_values('rank') #Sorting by rank
```

```python
rank_df.to_csv('rank.csv', index = False)
```

```python
cse_df = rank_df[
    rank_df["program"].str.contains("cs", case=False, na=False)
].copy()
cse_df["year_rank"] = (
    rank_df["cgpa"]
    .rank(method="dense", ascending=False)
    .astype(int)
) #Creating a separate file for CSE branch
```

```python
cse_df['rank'] = cse_df['cgpa'].rank(method='dense', ascending=False).astype(int) #Created another column for branch rank
```

```python
cse_df.to_csv("cse_rank2.csv", index = False)
```
